```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete_df = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

school_data_complete_df.head()
```

```python
total_schools = len(school_data_complete_df["school_name"].unique())

total_schools
```

```python
total_students = school_data_complete_df["Student ID"].count()

total_students
```

```python
grouped_schools_df = school_data_complete_df.groupby("school_name")

#grouped_schools_df[["School ID", "budget"]].max()
```

```python
total_budget = sum(grouped_schools_df["budget"].max())

total_budget
```

```python
average_math_score = school_data_complete_df["math_score"].mean()

average_math_score
```

```python
average_reading_score = school_data_complete_df["reading_score"].mean()

average_reading_score
```

```python
passing_math_students_df = school_data_complete_df.loc[school_data_complete_df["math_score"] >= 70]

passing_math_students_df

len(passing_math_students_df)
```

```python
math_pass_rate = (len(passing_math_students_df) / total_students) * 100

math_pass_rate
```

```python
passing_reading_students_df = school_data_complete_df.loc[school_data_complete_df["reading_score"] >= 70]

passing_reading_students_df

len(passing_reading_students_df)
```

```python
reading_pass_rate = (len(passing_reading_students_df) / total_students)* 100

reading_pass_rate
```

```python
overall_passing_students_df = school_data_complete_df.loc[(school_data_complete_df["math_score"] >= 70) & (school_data_complete_df["reading_score"] >= 70)]

overall_pass_rate = (len(overall_passing_students_df) / total_students) * 100

overall_pass_rate
```

```python
district_summary_dict = {"Total Schools" : total_schools,
                         "Total Students" : total_students,
                         "Total Budget" : total_budget,
                         "Average Math Score" : average_math_score,
                         "Average Reading Score" : average_reading_score,
                         "% Passing Math" : math_pass_rate,
                         "% Passing Reading" : reading_pass_rate, 
                         "% Overall Passing" : overall_pass_rate}

district_summary_dict
```

```python
district_summary_df = pd.DataFrame(district_summary_dict, index = [0])

district_summary_df
```

```python
school_summary_dict = {"School Type" : type_by_school,
                         "Total Students" : students_by_school,
                         "Total School Budget" : budget_by_school,
                         "Per Student Budget" : per_student_budget,
                         "Average Math Score" : math_score_by_school,
                         "Average Reading Score" : reading_score_by_school,
                         "% Passing Math" : per_school_passing_math,
                         "% Passing Reading" : per_school_passing_reading, 
                         "% Overall Passing" : per_school_passing_overall}
```

```python
ninth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "9th")]
tenth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "10th")]
eleventh_graders = school_data_complete_df[(school_data_complete_df["grade"] == "11th")]
twelfth_graders = school_data_complete_df[(school_data_complete_df["grade"] == "12th")]

ninth_grade_math_scores = ninth_graders.groupby(["school_name"]).mean()["math_score"]
tenth_grade_math_scores = tenth_graders.groupby(["school_name"]).mean()["math_score"]
eleventh_grade_math_scores = eleventh_graders.groupby(["school_name"]).mean()["math_score"]
twelfth_grade_math_scores = twelfth_graders.groupby(["school_name"]).mean()["math_score"]
```
